```python
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict

plt.rc('text', usetex=True)
plt.rc('font', family='serif')
FONTSIZE = 15
```

## Oppgave 2

```python
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']


x = np.linspace(-20, 20, num = 1000)
likning1 = (7 - x*x)/3
likning2 = 3*x - 1

fig = plt.figure(figsize = (4, 3))
ax = fig.add_subplot(1, 1, 1)

plt.title("Skisse av likningssystemet", fontsize = FONTSIZE)
plt.plot(x, likning1, linewidth=2, label = 'Likning 1')
plt.plot(x, likning2, linewidth=2, label = 'Likning 2')
plt.grid(True, ls='--')


plt.legend(loc = 'best', fontsize = FONTSIZE * 0.75)

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')


h = 10
plt.xlim([-15, 5])
plt.ylim([-35, 10])
plt.tight_layout()
plt.savefig('../figs/del1_oppg2.pdf')
plt.show()
```

## Del 1- oppgave 7

```python
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']


x = np.linspace(-20, 20, num = 2000)
f = (2*x - 1) / (x + 2)
#f_deriv = (x+2)**2 *(1-x)


fig = plt.figure(figsize = (4, 3))
ax = fig.add_subplot(1, 1, 1)

plt.title("Skisse av $f(x) = (2x - 1) / (x + 2) $", fontsize = FONTSIZE)
plt.plot(x[x < -2], f[x < -2], linewidth=2, label = '$f(x)$', color=colors[0])
plt.plot(x[x > -2], f[x > -2], linewidth=2, color=colors[0])

plt.plot([-20, 20], [2, 2], linewidth=2, label = '$y = 2$', color=colors[1])
plt.plot([-2, -2], [-20, 20], linewidth=2, label = '$x = -2$', color=colors[2])
#plt.plot(x, f_deriv, '--', linewidth=2, label = "$f'(x)$")
plt.grid(True, ls='--')


#plt.xlabel('$x$', fontsize = FONTSIZE)
#plt.ylabel('$y$', fontsize = FONTSIZE)
plt.legend(loc = 'best', fontsize = FONTSIZE * 0.75)

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')


h = 8
plt.xlim([-2 - h, -2 + h])
plt.ylim([2 - h, 2 + h])
plt.tight_layout()
plt.savefig('../figs/del1_oppg7.pdf')
plt.show()
```

## Del 1 - oppgave 9

```python
x = np.linspace(-6, 6, num = 2000)
f = -(x**4 /4 + x**3 - 4*x) + 2
f_deriv = (x+2)**2 *(1-x)


fig = plt.figure(figsize = (4, 3))
ax = fig.add_subplot(1, 1, 1)

plt.title("Skisse av $f'(x) = (x+2)^2(1-x)$", fontsize = FONTSIZE)
plt.plot(x, f, linewidth=2, label = '$f(x)$')
plt.plot(x, f_deriv, '--', linewidth=2, label = "$f'(x)$")
plt.grid(True, ls='--')


#plt.xlabel('$x$', fontsize = FONTSIZE)
#plt.ylabel('$y$', fontsize = FONTSIZE)
plt.legend(loc = 'best', fontsize = FONTSIZE * 0.75)

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')


plt.xlim([-5, 5])
plt.ylim([-4, 7])
plt.tight_layout()
plt.savefig('../figs/del1_oppg9b.pdf')
plt.show()
```

## Del 2 - oppgave 2

```python
from scipy import stats

n = list(range(115, 135 + 1))
p = 0.94

P_leq_116 = [stats.binom(k, p).cdf(116) for k in n]
P_leq_116

fig = plt.figure(figsize = (8, 3))
ax = fig.add_subplot(1, 2, 1)

plt.title("$P(X_n \leq 116)$  for ulike verdier av $n$", fontsize = FONTSIZE)
plt.plot(n, P_leq_116, '-o',linewidth=2, markersize=3, label = '$P(X_n \leq 116)$')
plt.grid(True, ls='--')


plt.xlabel('$n$', fontsize = FONTSIZE)
plt.ylabel('Sannsynlighet', fontsize = FONTSIZE)
plt.legend(loc = 'best', fontsize = FONTSIZE * 0.75)

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
#ax.spines['left'].set_position('center')
#ax.spines['bottom'].set_position('zero')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')


#plt.xlim([-5, 5])
#plt.ylim([-4, 7])
plt.tight_layout()



ax = fig.add_subplot(1, 2, 2)

plt.title("Samme plott, men zoomet inn", fontsize = FONTSIZE)
plt.plot(n, P_leq_116, '-o', linewidth=2.5, label = '$P(X_n \leq 116)$')
plt.grid(True, ls='--')


plt.xlabel('$n$', fontsize = FONTSIZE)
#plt.ylabel('Sannsynlighet', fontsize = FONTSIZE)
plt.legend(loc = 'best', fontsize = FONTSIZE * 0.75)

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
#ax.spines['left'].set_position('center')
#ax.spines['bottom'].set_position('zero')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')


plt.xlim([116, 120])
plt.ylim([0.9, 1.02])
plt.tight_layout()




plt.savefig('../figs/del2_oppg2.pdf')
plt.show()
```

## Oppgave 3

```python
x = np.linspace(-6, 6, num = 2000)
P = x**3 - 3*x**2 - 13*x + 15
P_deriv = 3*x**2 - 6*x**1 - 13

fig = plt.figure(figsize = (4, 3))
ax = fig.add_subplot(1, 1, 1)

plt.title('Skisse av $P(x) = x^3 - 3x^2 - 13x + 15$', fontsize = FONTSIZE)
plt.plot(x, P, linewidth=2, label = '$P(x)$')
plt.plot(x, P_deriv, '--', linewidth=1.5, label = "$P'(x)$")
plt.grid(True, ls='--')


#plt.xlabel('$x$', fontsize = FONTSIZE)
#plt.ylabel('$y$', fontsize = FONTSIZE)
plt.legend(loc = 'best', fontsize = FONTSIZE * 0.75)

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')


plt.xlim([-4, 6])
plt.ylim([-30, 30])
plt.tight_layout()
plt.savefig('../figs/oppg_3.pdf')
plt.show()
```
