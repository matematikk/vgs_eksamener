```python
import numpy as np
import matplotlib.pyplot as plt
```

# Oppgave 6

```python

x = np.linspace(-5,5,num=1000)
f = (np.exp(2*x) - 4*np.exp(x) + 3)

fig = plt.figure(figsize = (4,3))
ax = fig.add_subplot(1,1,1)
plt.title('Skisse av grafen $f(x) = e^{2x} - 4e^x + 3$')
plt.plot(x,f,linewidth = 2,label='$f(x)$')
plt.grid(True,ls = '--')
plt.legend()


#Move the axes so that they show wanted parts of the function
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

#Remove the upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.xlim([-3,1.5])
plt.ylim([-2,5])
plt.tight_layout()
plt.savefig('../figs/del1_oppg6d.pdf')
plt.show()
```
