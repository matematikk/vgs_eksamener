```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Del 1 - Oppgave 5

```python
plt.figure(figsize=(7,3))
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

x = np.linspace(-10, 10, num=1000)
f_d =(x-6)*(x+2)/4
f_dd = 0.5 * x - 1
f = x**3/12 - x**2/2 - 3*x + (23 - 32/6)
t = -3*x + 15

lw = 2
plt.plot(x, f, label=r"$f(x)$", lw=lw)
plt.plot(x, f_d, label=r"$f \, '(x)$", lw=lw)
plt.plot(x, f_dd, label=r"$f \, ''(x)$", lw=lw)
plt.plot(x, t, '--', label=r"$y = -3x + 15$", lw=lw)

plt.grid(True, linestyle='--', alpha=0.5)
plt.xlim([-8, 8])
plt.ylim([-8, 22])
plt.legend(loc='best', fontsize=11)
plt.savefig(r'../figs/del1_oppg5.pdf')
plt.show()
```

## Del 2 - Oppgave 2

```python
plt.figure(figsize=(7,3))
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

from scipy.stats import linregress
prop_cycle = plt.rcParams['axes.prop_cycle'].by_key()['color']

x = np.array([1990, 1995, 2000, 2005, 2010]) - 1990
y = np.array([2, 16, 41, 63, 105])

plt.scatter(x, y, zorder=25, label='Antall dde jerv')

slope, intercept, r_value, p_value, std_err = linregress(x, y)
x_cont = np.linspace(np.min(x) - 1, np.max(x) + 1)
plt.plot(x_cont, intercept + x_cont*slope, '--', color=prop_cycle[1], lw=3, zorder=25, 
         label=f'Liner modell $y=5.06x - 5.2$')

plt.ylabel('Antall dde jerv')
plt.xlabel('Antall r etter 1990')
plt.xticks(x)
#plt.ylim([-40, 110])
plt.legend(loc=(0.15, 0.75))
plt.grid(True, linestyle='--', alpha=0.5, zorder=-15)
plt.savefig(r'../figs/del2_oppg2.pdf')
plt.show()
```

## Del 2 - Oppgave 2 c)

```python
plt.figure(figsize=(7, 3))
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

from scipy.stats import linregress
prop_cycle = plt.rcParams['axes.prop_cycle'].by_key()['color']

x_smooth = np.linspace(0, 10)
y = 5.06 * x_smooth - 5.2

#plt.scatter(x, y, zorder=25, label='Antall dde jerv')


plt.plot(x_smooth, y, '-', color=prop_cycle[0], lw=2, zorder=25, 
         #label=f'Liner modell $y=5.06x - 5.2$'
        )

plt.fill_between(x_smooth, y, '-', color=prop_cycle[0], lw=1, zorder=25, alpha=0.2,
         label=r'Integralet $ \int_0^{10} f(x) \,  dx$'
        )


x_vals = np.arange(0, 10 + 1)
plt.bar(x_vals, 5.06 * x_vals - 5.2, width=1, edgecolor=prop_cycle[1], fc=(0, 0, 1, 0),
       label=r'Summen $ \sum_{x=0}^{10} f(x)$', lw=1.5)

plt.ylabel('Antall dde jerv')
plt.xlabel('Antall r etter 1990')
plt.xticks(np.arange(0, 10+1))
plt.yticks(np.arange(-50, 100, 10))
#plt.ylim([-40, 110])
plt.legend(loc=(0.15, 0.55))
plt.grid(True, linestyle='--', alpha=0.5, zorder=-15)
plt.xlim([-1.5, 11])
plt.ylim([-10, 50])
plt.savefig(r'../figs/del2_oppg2_c.pdf')
plt.show()
```

## Del 2 - Oppgave 5

```python
from scipy.stats import binom, norm
prop_cycle = plt.rcParams['axes.prop_cycle'].by_key()['color']

n = 1000
p = 0.1

plt.figure(figsize=(7,3))

# Binomisk fordelt
X = binom(n=n, p=p)
x = np.arange(75, 140 + 1)
y = X.pmf(x)
plt.plot(x, y, lw=3,label='Fordelingen til $X \sim binom$', zorder=20)

# Normalfordelt
X_norm = norm(loc = n*p, scale= np.sqrt(n * p * (1-p)))
x = np.arange(75, 140 + 1)
y = X_norm.pdf(x)
plt.plot(x, y, '--', lw=2, label='Fordelingen til $X \sim normal$', zorder=25)

x = np.arange(110, 140 + 1)
y = X.pmf(x)
plt.fill_between(x, y, color = prop_cycle[4], label='$P(X \geq 110)$', zorder=15)
plt.legend(loc='upper right')
plt.grid(True, alpha=0.5, zorder=-15)
plt.savefig(r'../figs/del2_oppg5.pdf')
plt.show()
```
